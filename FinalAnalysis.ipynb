```python
import os
import pandas as pd

# Specify the directory where your CSV files are located
directory_path = '/content/gdrive/MyDrive/urbcomp/Crime/datasets/'

# Create an empty dictionary to store DataFrames for each year
crime_data_by_year = {}

# Iterate through each CSV file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        # Read the CSV file into a DataFrame
        year = filename.split('_')[-1].split('.')[0]  # Extract the year from the filename
        file_path = os.path.join(directory_path, filename)
        crime_data_for_year = pd.read_csv(file_path)

        # Add the data for the current year to the dictionary
        crime_data_by_year[int(year)] = crime_data_for_year

# Now, crime_data_by_year is a dictionary where each key is a year, and the corresponding value is a DataFrame for that year
# Access a specific year's DataFrame using crime_data_by_year[desired_year]
# For example, to access the DataFrame for the year 2022: crime_data_by_year[2022]
```

```python
violent_crimes = ['BATTERY', 'ASSAULT', 'CRIMINAL SEXUAL ASSAULT', 'HOMICIDE', 'KIDNAPPING', 'CRIM SEXUAL ASSAULT']
narc = ['NARCOTICS']
prostitution = ['PROSTITUTION']
theft = ['THEFT']
burglary = ['BURGLARY']

crime_violent_by_year = {}
crime_nonviolent_by_year = {}

for year, crime_df in crime_data_by_year.items():
    violent_crime_df = crime_df[crime_df['Primary Type'].isin(violent_crimes)]
    nonviolent_crime_df = crime_df[~crime_df['Primary Type'].isin(violent_crimes)]

    crime_violent_by_year[year] = violent_crime_df
    crime_nonviolent_by_year[year] = nonviolent_crime_df
```
